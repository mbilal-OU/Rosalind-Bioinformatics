```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df =sns.load_dataset('tips')
df.head()
```

```python
# slelect the feature column
x=df[['total_bill']]
# select target column
y=df['tip']
```

```python
# load model
model=LinearRegression()
# fit the mode
model.fit(x,y)
```

```python
# lets predict the tip for an unknown bill
# total bill=50
model.predict([[50]])
```

```python
# create a regression line plot
sns.lmplot(x='total_bill',y='tip',data=df)
```
